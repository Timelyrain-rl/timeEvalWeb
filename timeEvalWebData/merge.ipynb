```python
import pandas as pd
import numpy as np
from datetime import datetime

df = pd.read_csv('month_3.csv')

df['每日对应秒序号'] = df['秒序号']

try:
    df['记录仪时间'] = pd.to_datetime(df['记录仪时间'])
except:
    # 格式为 yyyy-mm-dd HH:MM:SS
    try:
        df['记录仪时间'] = pd.to_datetime(df['记录仪时间'], format='%Y-%m-%d %H:%M:%S')
    except:
        print("无法转换记录仪时间列，请检查日期格式并修改代码中的format参数")

df['年'] = df['记录仪时间'].dt.year
df['月'] = df['记录仪时间'].dt.month
df['日'] = df['记录仪时间'].dt.day
df['小时'] = df['记录仪时间'].dt.hour
df['分钟'] = df['记录仪时间'].dt.minute

df['日期'] = df['记录仪时间'].dt.date

df['每日对应秒序号'] = df.groupby('日期').cumcount()

df.to_csv('month_3_processed.csv', index=False)

print("处理完成！新文件已保存为month_3_processed.csv")
```
